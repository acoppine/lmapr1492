## Détermination des 3 premiers pics du diffractogramme du cristal de C

```python
from pymatgen.ext.matproj  import MPRester
import matplotlib.pyplot as plt
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

### Téléchargement du fichier CIF du cristal de C

```python
with MPRester("MoXgF85AP31iKfy0") as m:
    structure =  m.get_structure_by_material_id("mp-169")
    
xrd = XRDCalculator(1.54060)
pattern = xrd.get_pattern(structure)
```

### Détermination des 3 premiers pics

Dans ce notebook, nous allons déterminer les 3 premiers pics du diffractogramme du cristal de C. Nous pouvons faire cela en étudiant la diffraction d'un faisceau de lumière infrarouge.Nous devons considérer un rayonnement X incident Cu Kα de longueur d'onde $\lambda$ = 1.54060 Å. Cela se fait en prenant plusieurs (beaucoup) fois un certain angle 2$\theta$. Cela nous permet d'obtenir le spectre de la diffraction. 
$\newline$ Voici donc les résultats obtenus :


#### Premier pic

```python
print("Angle 2θ : "+str(pattern.x[0])+"°")
print("Intensité du pic (qui est un pourcentage par rapport à l'intensité la plus forte) : "+ str(pattern.y[0])+" %")
print("Indices de Miller et leur multiplicité : "+str(pattern.hkls[0][0]["hkl"])+ " de multiplicité "+str(pattern.hkls[0][0]["multiplicity"]))
print("Finalement, on a la distance entre les plans (en effet, on peut considérer notre cristal comme une série de plans) noté d. Et  donc d : "+str(pattern.d_hkls[0])+" Å")
```

#### Deuxième pic

```python
print("Angle 2θ : "+str(pattern.x[1])+"°")
print("Intensité du pic (qui est un pourcentage par rapport à l'intensité la plus forte) : "+ str(pattern.y[1])+" %")
print("Indices de Miller et leur multiplicité : "+str(pattern.hkls[1][0]["hkl"])+ " de multiplicité "+str(pattern.hkls[1][0]["multiplicity"]))
print("Finalement, on a la distance entre les plans (en effet, on peut considérer notre cristal comme une série de plans) noté d. Et  donc d : "+str(pattern.d_hkls[1])+" Å")
```

#### Troisième pic

```python
print("Angle 2θ : "+str(pattern.x[2])+"°")
print("Intensité du pic (qui est un pourcentage par rapport à l'intensité la plus forte) : "+ str(pattern.y[2])+" %")
print("Indices de Miller et leur multiplicité : "+str(pattern.hkls[2][0]["hkl"])+ " de multiplicité "+str(pattern.hkls[2][0]["multiplicity"]))
print("Finalement, on a la distance entre les plans (en effet, on peut considérer notre cristal comme une série de plans) noté d. Et  donc d : "+str(pattern.d_hkls[2])+" Å")
```
